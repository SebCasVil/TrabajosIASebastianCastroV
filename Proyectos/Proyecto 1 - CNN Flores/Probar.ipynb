# Probar CNN en múltiples carpetas
* Lo primero es cargar el modelo ya entrenado
* Después se abre una carpeta que contiene múltiples carpetas y se itera sobre cada una
* Se imprime el nombre de la carpeta y se itera sobre las imagenes en cad carpeta
* Se obtiene el arreglo de información de la imagen y se predice con el modelo
* Se imprime la predicción

```python
import tensorflow as tf
import os
from tensorflow.keras.preprocessing import image
import numpy as np

modelo = 'C:/Users/kalan/Desktop/IA/Flores2/Generar/modeloDef5/'
model = tf.keras.models.load_model(modelo)
carpetaPruebas = 'C:/Users/kalan/Desktop/IA/Flores2/Generar/Pruebas4/'
carpetas = [f.path for f in os.scandir(carpetaPruebas) if f.is_dir()]
target_size = (200, 200)
for carpeta in carpetas:
    carpetaNombre = os.path.basename(carpeta)
    print(carpetaNombre)

    imagenes = [os.path.join(carpeta, f) for f in os.listdir(carpeta) if f.endswith('.jpg')]

    for imagen in imagenes:
        img = image.load_img(imagen, target_size=target_size)
        img_array = image.img_to_array(img)
        img_array = np.expand_dims(img_array, axis=0)

        predictions = model.predict(img_array)
        print("Prediccion para", os.path.basename(imagen), ":", predictions)

        class_labels = ["DienteAmarillo", "Girasol", "Lantana", "Lirio", "Vinca"]
        predicted_class_indices = np.argmax(predictions, axis=1)
        predicted_labels = [class_labels[i] for i in predicted_class_indices]

        print("Predicted class:", predicted_labels[0])

# 0 DienteAmarillo
# 1 Girasol
# 2 Lantana
# 3 Lirio
# 4 Vinca
```
